```python
# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np

# Configuración para mostrar todas las columnas
pd.set_option('display.max_columns', None)
```

```python
# Cargar el archivo CSV
df = pd.read_csv('Olympic_Athlete_Event_Details.csv', encoding='utf-8')

# Mostrar las primeras filas para inspección
df.head()
```

```python
# Información general del DataFrame
df.info()

# Estadísticas descriptivas
df.describe(include='all')

# Verificar valores nulos
df.isnull().sum()
```

```python
# LImpieza de data
# Verificar qué columnas podríamos eliminar 
columns_to_drop = []  # Ajustar según necesidad

if columns_to_drop:
    df = df.drop(columns=columns_to_drop)
```

```python
# Limpiar columnas de texto (eliminar espacios extras, etc.)
text_cols = ['edition', 'country_noc', 'sport', 'event', 'athlete', 'medal']
for col in text_cols:
    df[col] = df[col].str.strip()
```
